```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
```

```python
dataset_tn = pd.read_csv('Adult2_Training.csv')
dataset_tt=pd.read_csv('Adult2_Test.csv')
dataset_tn.head()
```

```python
dataset_tt.head()
```

```python
# create dataset
df_tn = pd.DataFrame(dataset_tn)
df_tt = pd.DataFrame(dataset_tt)
        
# create dummy variables
dm_tn=pd.get_dummies(df_tn)
dm_tt=pd.get_dummies(df_tt)
```

```python
dm_tn.head()
```

```python
dm_tt.head()
```

```python
dm_tn= dm_tn.drop(columns=['Income_<=50K','Capital Gains or Losses_F'], axis=0)
```

```python
dm_tt= dm_tt.drop(columns=['Income_<=50K','Capital Gains or Losses_F'], axis=0)
```

```python
dm_tn.head()
```

```python
dm_tt.head()
```

```python
# Assign every variable except Income_>50Kto be X
dfX_tn = dm_tn.loc[:, dm_tn.columns != 'Income_>50K']

```

```python
# Assign every variable except Income_>50Kto be X
dfX_tt = dm_tt.loc[:, dm_tt.columns != 'Income_>50K']
```

```python
# Assign Competitive? to be Y
dfY_tn = dm_tn['Income_>50K']
dfY_tt = dm_tt['Income_>50K']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
tree = DecisionTreeClassifier(min_samples_split=50)
tree.fit(dfX_tn, dfY_tn)
```

```python
Y_predicted = tree.predict(dfX_tt)
Y_predicted
```

```python
tree.score(dfX_tt,dfY_tt)
```

```python
confusion_matrix = confusion_matrix(dfY_tt, Y_predicted)
```

```python
gl_cf=confusion_matrix
gl_cf
```

```python
# Assign every variable except Income_>50Kto be X
dfX_tn = df_tn.loc[:, df_tn.columns != 'Income_>50K']
# Assign every variable except Income_>50Kto be X
dfX_tt = df_tt.loc[:, df_tt.columns != 'Income_>50K']
# Assign Income_>50K to be Y
dfY_tn = df_tn['Income_>50K']
dfY_tt = df_tt['Income_>50K']
tree = DecisionTreeClassifier(min_samples_split=50)
tree.fit(dfX_tn, dfY_tn)
```

```python
nc_cf=confusion_matrix
nc_cf
```

```python
#Combined Confusion matrix(CapGains+Losses --> reported as well as non reported)
cc_confusion_matrix=nc_cf+c_cf
```

```python
cc_confusion_matrix
```

```python
def sum1(input):
    return sum(map(sum, input))
  
```

```python
cc_accuracy=(cc_confusion_matrix[0][0]+ cc_confusion_matrix[1][1])/sum1(cc_confusion_matrix)
```
